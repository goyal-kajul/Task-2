# Data set - Delhi weather conditions

- Clean the data. Detect and remove outliers.
- Make a plot of different numeric features like rain, snow, dewptm, etc. over the duration of a month or year or day.
 In each subplot, 
 y-axis: any numeric feature
 x-axis: time (in case of month, it has range 1-12)
- Make pie charts for different weather conditions in a month.

```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import datetime
from pylab import rcParams
%matplotlib inline
```

```python
df=pd.read_csv("testset.csv", parse_dates=['datetime_utc'])
```

```python
df.drop(['_precipm'],inplace=True,axis=1)
```

```python
df['_conds'].fillna("Unknown",inplace=True)
df['_dewptm'].fillna(df['_dewptm'].mean(),inplace=True)
df['_heatindexm'].fillna(df['_heatindexm'].mean(),inplace=True)
df['_hum'].fillna(df['_hum'].mean(),inplace=True)
df['_pressurem'].fillna(df['_pressurem'].mean(),inplace=True)
df['_tempm'].fillna(df['_tempm'].mean(),inplace=True)
df['_vism'].fillna(df['_vism'].mean(),inplace=True)
df['_wdird'].fillna(df['_wdird'].mean(),inplace=True)
df['_wdire'].fillna("none",inplace=True)
df['_wgustm'].fillna(df['_wgustm'].mean(),inplace=True)
df['_windchillm'].fillna(df['_windchillm'].mean(),inplace=True)
df['_wspdm'].fillna(df['_wspdm'].mean(),inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['datetime_utc'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['hour'] = df['Date'].dt.hour
```

```python
fig,axes= plt.subplots(nrows=4,ncols=4,figsize=(25,25))
month_humi = df.groupby(df.month).mean()
month_humi._dewptm.plot(ax=axes[0,0],title='dewptm')
month_humi._fog.plot(ax=axes[0,1],title='fog')
month_humi._hail.plot(ax=axes[0,2],title='hail')
month_humi._heatindexm.plot(ax=axes[0,3],title='heatindexm')
month_humi._hum.plot(ax=axes[1,0],title='hum')
month_humi._pressurem.plot(ax=axes[1,1],title='pressurem')
month_humi._rain.plot(ax=axes[1,2],title='rain')
month_humi._snow.plot(ax=axes[1,3],title='snow')
month_humi._tempm.plot(ax=axes[2,0],title='temptm')
month_humi._thunder.plot(ax=axes[2,1],title='thunder')
month_humi._tornado.plot(ax=axes[2,2],title='tornado')
month_humi._vism.plot(ax=axes[2,3],title='vism')
month_humi._wdird.plot(ax=axes[3,0],title='wdird')
month_humi._wgustm.plot(ax=axes[3,1],title='wgustm')
month_humi._windchillm.plot(ax=axes[3,2],title='windchillm')
month_humi._wspdm.plot(ax=axes[3,3],title='wspdm')
```
